```python
###R1 - import data.csv

df = pd.read_csv("data.csv")
#print(df.to_string())
df
```

```python
###R2 - drop all column that has null

df = df.dropna()
df
```

```python
###R3 - check the data types of all features
df.info()
```

```python
###R4 – count all variables in features that is not numeric.

Count_repay_fail = df['repay_fail'].value_counts()
print(Count_repay_fail, '\n')

# Count the occurrences of variable in various features
term = df['term'].value_counts()
employment_length = df['employment_length'].value_counts()
home_ownership = df['home_ownership'].value_counts()
verification_status = df['verification_status'].value_counts()
loan_status = df['loan_status'].value_counts()
purpose = df['purpose'].value_counts()
revolving_utillization = df['revolving_utillization'].value_counts()

print(term, "\n")
print(employment_length, "\n")
print(home_ownership, "\n")
print(verification_status, "\n")
print(loan_status, "\n")
print(purpose, "\n")
print(revolving_utillization, "\n")
```

```python
###R5 - take out any string or symbol in features & convert string variables to numeric variables & convert the features data type to int/float

df = df.copy()

#take out word month in feature term and convert to int
df.loc[:, "term"] = df["term"].str.rstrip("months")
df["term"] = df["term"].astype(int)

#take out symbol in feature employment_length and convert to int
df.loc[:, "employment_length"] = df["employment_length"].str.lstrip("<")
df.loc[:, "employment_length"] = df["employment_length"].str.rstrip("+ years")
df["employment_length"] = df["employment_length"].astype(int)

#change string variable to numeric variable in feature home_ownership and convert feature to int
df.loc[df['home_ownership'] == 'RENT', 'home_ownership'] = 4
df.loc[df['home_ownership'] == 'MORTGAGE', 'home_ownership'] = 0
df.loc[df['home_ownership'] == 'OWN', 'home_ownership'] = 3
df.loc[df['home_ownership'] == 'OTHER', 'home_ownership'] = 2
df.loc[df['home_ownership'] == 'NONE', 'home_ownership'] = 1
df["home_ownership"] = df["home_ownership"].astype(int)

#change string variable to numeric variable in feature verification_status and convert feature to int
df.loc[df["verification_status"] == 'Not Verified', "verification_status"] = 1
df.loc[df["verification_status"] == 'Verified', "verification_status"] = 0
df.loc[df["verification_status"] == 'Source Verified', "verification_status"] = 2
df["verification_status"] = df["verification_status"].astype(int)

#change string variable to numeric variable in feature loan_status and convert feature to int
df.loc[df["loan_status"] == 'Fully Paid', "loan_status"] =5
df.loc[df["loan_status"] == 'Charged Off', "loan_status"] =0
df.loc[df["loan_status"] == 'Does not meet the credit policy. Status:Fully Paid', "loan_status"] =4
df.loc[df["loan_status"] == 'Current', "loan_status"] =1
df.loc[df["loan_status"] == 'Does not meet the credit policy. Status:Charged Off', "loan_status"] =3
df.loc[df["loan_status"] == 'Late (31-120 days)', "loan_status"] =8
df.loc[df["loan_status"] == 'In Grace Period', "loan_status"] =6
df.loc[df["loan_status"] == 'Late (16-30 days)', "loan_status"] =7
df.loc[df["loan_status"] == 'Default', "loan_status"] =2
df["loan_status"] = df["loan_status"].astype(int)

#change string variable to numeric variable in feature purpose and convert feature to int
df.loc[df["purpose"] == 'debt_consolidation', "purpose"] =4
df.loc[df["purpose"] == 'credit_card', "purpose"] =1
df.loc[df["purpose"] == 'other', "purpose"] =2
df.loc[df["purpose"] == 'home_improvement', "purpose"] =3
df.loc[df["purpose"] == 'major_purchase', "purpose"] =7
df.loc[df["purpose"] == 'small_business', "purpose"] =0
df.loc[df["purpose"] == 'car', "purpose"] =13
df.loc[df["purpose"] == 'wedding', "purpose"] =10
df.loc[df["purpose"] == 'medical', "purpose"] =12
df.loc[df["purpose"] == 'moving', "purpose"] =9
df.loc[df["purpose"] == 'educational', "purpose"] =6
df.loc[df["purpose"] == 'house', "purpose"] =5
df.loc[df["purpose"] == 'vacation', "purpose"] =11
df.loc[df["purpose"] == 'renewable_energy', "purpose"] =8
df["purpose"] = df["purpose"].astype(int)      
         
#take out symbol in feature revolving_utillization and convert to float
df.loc[:, 'revolving_utillization'] = df['revolving_utillization'].str.rstrip("%")
df["revolving_utillization"] = df["revolving_utillization"].astype(np.float64)

df
```
